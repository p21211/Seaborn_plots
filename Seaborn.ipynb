```python
# solution 1
# five plots by seaborn are:  Line plot,Countplot,Bar chart,Pairplot,Scatter plot
# its usage
# Line plot:Line plots show how data is spread out.
# Countplot:A univariate plot that shows the comparison of different groups in categorical variables. 
# Bar chart:Bar graphs are used to compare data. 
# Pairplot:allows us to plot pairwise relationships between variables within a dataset.
# Scatter plot:a diagram where each value in the data set is represented by a dot. 
```

```python
# solution 2:
import seaborn as sns
fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot(data=fmri, x=fmri.timepoint, y=fmri.signal, hue=fmri.event, style=fmri.region)
```

```python
# solution 3:
import seaborn as sns
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

```python
sns.boxplot(data=titanic,  x=titanic.pclass,y=titanic.age)
```

```python
sns.boxplot(data=titanic,  x=titanic.pclass,y=titanic.fare)
```
